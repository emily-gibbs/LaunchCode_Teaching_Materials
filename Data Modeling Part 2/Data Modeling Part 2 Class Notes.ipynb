**Machine Learning** is the practice of using mathematical models to identify patterns within data and use those patterns to make predictions on data that was not used for training. This is all done without having to explicitly program the rules used for making those predictions. 

# What Type of Data Does ML Work With?

So many different kinds! But they all have something in common: independent and dependent variables.

* **Independent Variables** : These are variables that are 'features' of your particular problem. These are what a model takes in as input. For instance, if you are comparing houses some of the 'features' of the houses could be the square footage, age, number of rooms, type of flooring, etc. The more truly independent these variables are (each giving you new information that is not contained in another variable) the better, and all of the independent variables should contain enough information to determine the dependent variable. 

* **Dependent Variables** : These are the variables that depend on/are determined by the independent variables. Generally an ML model will only have a single dependent variable. It is the output of a model, and what we are trying to get a model to predict based on the independent variables that are input, so they are usually called the 'target variables'. For examples, if the independent variables are features of a house the dependent variable could be the price of the house, or the probability that it will sell. 

The features that go into the model are really important, and a lot of times you have to tweak them and put them into new forms that will work best for the model. This is called **feature engineering** and is a big part of Data Science.

### Regression Problem

This is the type of problem where the output is continuous. In our housing example, this would be the situation where you are trying to predict a house's selling price (dependent variable) based on the house features (independent variables).

Examples of regression problems:
* Predicting temperature
* Predicting revenue
* Predicting sales price

Popular regression algorithms include:
* Linear/Polynomial regression
* Support vector regression
* Regression trees

### Classification Problem

This is the type of problem where the output is discrete. Again with our housing example, if were were instead trying to predict whether the house is safe to live in or not it would be a classification problem. There are only two possible results (called the 'classes') - Safe and Not Safe. A classification problem must at least have two classes, but can have many more. You can think of classification models as working to determine the boundaries between the classes.

Some examples of classification problems:
* Classifying objects in an image

* Customer Churn prediction
* Document classification
* and so many more!

Popular classification algorithms include:
* Logistic regression

* Decision Trees and Random Forests

* Neural Networks

# Do You Have the Data You Need? 

Usually to train a ML model you give it a lot of examples of independent variables (features) with their correctly resulting dependent variable ('answer') so that the model can 'learn' by example. Data that includes the correct answer is what we call **labeled data**, and using labeled data to train a model is called **supervised learning**.

But what if you don't have data with the correct answer assigned to each example? You can still do ML, just a different kind called **unsupervised learning**.

<center><img src="https://cdn-images-1.medium.com/max/1600/1*6mPnd6tEA4EsYD1f72hGkA.png" width="800" height="800" />

The most popular form of unsupervised learning is clustering, where the model finds groups of data points that are close to each other, which are then used to classify new points based upon what cluster they fall into. 

# ML in Practice

When you train a model, you don't want to test how the model if performing using the same data that you used to train. Why? Because that's cheating!

<center><img src="https://media.tenor.com/images/d7f9586131867a7d56d442a2cd478ec0/tenor.gif" width="300" height="300" />

The model has already seen that data, so it has a higher chance of getting it right. To really know you well your model is going to generalize, you have to see how it works on data its never seen before. This is why we split out data into two sets: training data and testing data.

# A Classification Example

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
iris = load_iris()
print(iris.DESCR)
```

```python
iris_data = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
iris_data.head()
```

```python
target = pd.DataFrame(iris.target, columns=['iris_type'])
target.head()
```

```python
target['iris_type'].unique()
```

* 0 is Iris Setosa
* 1 is Iris Versicolour
* 2 is Iris Virginica

### Explore the Data

```python
combined_df = pd.concat([iris_data, target], axis=1)
combined_df
```

```python
combined_df.describe()
```

```python
sns.heatmap(combined_df.corr(), annot=True)
```

```python
sns.pairplot(combined_df, hue='iris_type', height=2, palette='colorblind')
```

### Make Train and Test Datasets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris_data, target, test_size=0.5, stratify=target)

# Stratifying makes sure that the different classes are spread out between the train and test data
```

### Train the Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, \
                            confusion_matrix, \
                            classification_report
import scikitplot as skplt
import matplotlib.pyplot as plt
```

```python
logistic_reg_model = LogisticRegression()
```

```python
# Train the model based on the training data
logistic_reg_model.fit(x_train, y_train.values.ravel())
```

```python
# Use the test data to see how well the model performs when predicting class on data
# it hasn't seen
y_pred = pd.Series(logistic_reg_model.predict(x_test))
print(f'Accuracy : {accuracy_score(y_test, y_pred)}')
cm = confusion_matrix(y_test, y_pred)
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=False)
```

### Another Type of Classification Model

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train, y_train.values.ravel())
y_pred = tree_model.predict(x_test)
print(f'Accuracy : {accuracy_score(y_test, y_pred)}')
cm = confusion_matrix(y_test, y_pred)
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=False)
```

```python
plt.figure(figsize = (15,12))
plot_tree(tree_model, feature_names = iris_data.columns, 
          class_names = ['setosa', 'versicolor', 'virginica'], filled = True)
plt.show()
```
